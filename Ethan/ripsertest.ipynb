```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.spatial.distance as d
import ripser
import tadasets
from persim import plot_diagrams

from persim.visuals import plot_a_bar

import gudhi
```

```python
# data = pd.read_csv("../datasets/CCMathTopologyScavengerHunt/points11.csv")
data = pd.read_csv("../datasets/PH-roadmap/data_sets/roadmap_datasets_distmat/dragon_vrip.ply.txt_1000_.txt_distmat.txt", header=None, delimiter=" ")
data = np.array(data)
print(data.shape)
```

```python
ripsOut = ripser.ripser(X=data, maxdim = 2, distance_matrix=True)
```

```python
from numpy import sort

print(ripsOut.keys())
ripsDGMS = ripsOut["dgms"]
plot_diagrams(ripsDGMS)

holeLifeTimes = ripsOut["dgms"][2][:, 1] - ripsOut["dgms"][2][:, 0]
sort(holeLifeTimes)

ripsOut["dgms"][:]
longestLife : float = np.max(ripsOut["dgms"][:][1])
lives = ripsOut["dgms"][:] # / longestLife
plot_diagrams(lives)
```
